```python
import matplotlib.pyplot as plt
```

```python
n_values = [100000, 200000, 300000, 400000, 500000]
selection_sort_times = [5.211, 21.289, 46.960, 85.38, 168.979]
insertion_sort_times = [2, 7.91, 16.12, 31.71, 65.48]
bubble_sort_times = [5.582, 21.914, 49.414, 96.936, 212.979]
```

```python
plt.plot(n_values, selection_sort_times, marker='o', label='Selection Sort')
plt.plot(n_values, insertion_sort_times, marker='o', label='Insertion Sort')
plt.plot(n_values, bubble_sort_times, marker='o', label='Bubble Sort')


plt.xlabel('Array Size (n)')
plt.ylabel('Time (seconds)')
plt.title('Sorting Algorithm Performance')
plt.legend()
plt.grid(True)

plt.show()
```
