```python
#Numpy
import numpy as np
list1=[1,2,3,4]
list2=[5,6,7,8]
list3=[20,5,2,0]
arr=np.array([list1,list2,list3])
print(arr)
Arr=arr.reshape(4,3)     # Reshapes arr according to given specifications.
print(Arr)
```

```python
#Numpy arrays are faster than tuples and lists.
t1=(1,2,3,4)
t2=(5,6,7,8)
arr1=np.array(t1)
print(arr1)
arr2=np.array(t2)
print(arr2)
arr3=np.array([t1,t2])
print(arr3)
arr4=np.array(t1+t2)
print(arr4)
print(arr3[1:2,:2])
print(arr3[:,1:3])

```

```python
arr5=np.arange(1,21).reshape(4,5)
print(arr5)
arr6=np.arange(1,20)
print(arr6)
```

```python
arr7=np.random.randint(0,50,4).reshape(2,2)    #a way to create arrays by randomly picking numbers and adding to a array.
print(arr7)                                    #(0,50,4) Here 4 means that 4 numbers are picked from 0 to 50 and added to array.
```
